#### Grafico de lineas: Total de Conexiones por Rango de Velocidad (2018 al 2022 - Q)

```python
# Calcula la cantidad de las conexiones agrupando por año y rango de velocidad
conexiones_por_año = internet_rangos_vel_prov_2018.groupby(['Año', "Trimestre"]).sum()

# Reinicia el índice para tener los años como columna
conexiones_por_año = conexiones_por_año.reset_index()

# Convierte el DataFrame en un formato largo para poder utilizar 'hue' en el gráfico de líneas
sumas_largo = pd.melt(conexiones_por_año, id_vars=['Año', "Trimestre"], var_name='Rango de Velocidad', value_name='Cantidad de Conexiones')


# Creo una nueva columna que concatene el año con el trimestre:
sumas_largo["Año_Trim"]= sumas_largo["Año"].astype(str) + "-" + sumas_largo["Trimestre"].astype(str)

# Crea el gráfico de líneas con 'hue' para los rangos de velocidad
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

sns.lineplot(data=sumas_largo, x="Año_Trim", y='Cantidad de Conexiones', hue='Rango de Velocidad')

# Establece los títulos y etiquetas de los ejes
plt.title('Total Conexiones por Rango de Velocidad (2018-2022 - Trimestral)')
plt.xlabel('Año-Trimestre')
plt.ylabel('Cantidad de Conexiones')

# Muestra el gráfico
plt.show()
```

#### Grafico de lineas: Mediana de Conexiones por Rango de Velocidad (2018 al 2022 - Q)

```python
# Calcula la cantidad de las conexiones agrupando por año y rango de velocidad
conexiones_por_año = internet_rangos_vel_prov_2018.groupby(['Año', "Trimestre"]).median()

# Reinicia el índice para tener los años como columna
conexiones_por_año = conexiones_por_año.reset_index()

# Convierte el DataFrame en un formato largo para poder utilizar 'hue' en el gráfico de líneas
sumas_largo = pd.melt(conexiones_por_año, id_vars=['Año', "Trimestre"], var_name='Rango de Velocidad', value_name='Cantidad de Conexiones')

# Creo una nueva columna que concatene el año con el trimestre:
sumas_largo["Año_Trim"]= sumas_largo["Año"].astype(str) + "-" + sumas_largo["Trimestre"].astype(str)

# Crea el gráfico de líneas con 'hue' para los rangos de velocidad
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

sns.lineplot(data=sumas_largo, x="Año_Trim", y='Cantidad de Conexiones', hue='Rango de Velocidad')

# Establece los títulos y etiquetas de los ejes
plt.title('Mediana de Conexiones por Rango de Velocidad (2018 al 2022 - Q)')
plt.xlabel('Año-Trimestre')
plt.ylabel('Cantidad de Conexiones')

# Muestra el gráfico
plt.show()
```

 * Durante el periodo comprendido entre 2018 y 2022, el rango de velocidad más común y con más accesos a conexiones en todas las provincias de Argentina fue de +1 a -6 Mbps.
 * A lo largo del tiempo, el rango de velocidad de +30 Mbps ha experimentado un aumento gradual, siendo inicialmente bajo en el pasado.
 * En la actualidad, el rango de velocidad de +30 Mbps supera tanto en el total de conexiones como en la mediana al rango de 1 a 6 Mbps. Indicando que Argentina esta apostando por mejorar las conexiones y velocidades del pais.
 * Se puede observar una relación inversamente proporcional entre el aumento de conexiones en el rango de velocidad de +30 Mbps y la disminución en el rango de velocidad de 1 a 6 Mbps.
 * Se observa una correlación entre las conexiones perdidas en el rango de velocidad de 1 a 6 Mbps y las conexiones incrementadas en el rango de velocidad de +30 Mbps.
 * En el gráfico, se puede notar un cruce y un cambio en la tendencia de ambas velocidades, lo cual sugiere un cambio en la preferencia de los usuarios hacia velocidades más altas.
 * La intersección de estás dos velocidades, en el primer grafico(Total), se da en el 4to trimestre del 2019, en donde la velocidad de +30 Mpbs supera a + 1 a 6 Mbps y en la grafica de la (mediana) se da en el 4to trimestre del 2021, 2 años despues. Esto podría sugerir que aunque esa velocidad viene en crecimiento, y supera a nivel general de Argentina a la velocidad de +1 a 6 Mbps, a nivel Provincial, se ha visto solamente despues de 2 años.

 * En general, las velocidades que vienen creciendo en cuanto a conexiones tanto en el total como en la mediana de las provincias, son +30 Mbps, +10 - 20 Mbps, +6 a 10 Mbps (Aunque decrecio en el ultimo trimestre del 2022) y +20 - 30 Mbps, lo que vuelve a indicar que se estan creando más conexiones enfocadas en mejores velocidades y se estan dejando de lado las conexiones de +1 a 6 Mbps.



#### Grafico de barras: Cantidad de Conexiones por Provincia (2018 al 2022 - Q1)

```python
# Filtrar los datos solo para el primer trimestre de cada año
internet_rangos_vel_prov_pri_Trim = internet_rangos_vel_prov_2018[internet_rangos_vel_prov_2018['Trimestre'] == 1]

# Crear la columna 'Periodo' concatenando el año y el trimestre
internet_rangos_vel_prov_pri_Trim['Periodo'] = internet_rangos_vel_prov_pri_Trim['Año'].astype(str) + '-Q' + internet_rangos_vel_prov_pri_Trim['Trimestre'].astype(str)
```

```python
# Calcular las conexiones totales por provincia y periodo
conexiones_totales_provincia = internet_rangos_vel_prov_pri_Trim.groupby(['Provincia', 'Año', 'Periodo'])['Total'].sum().reset_index()
```

```python
# Ordenar las provincias por la cantidad total de conexiones en orden descendente
provincias_ordenadas = conexiones_totales_provincia.groupby('Provincia')['Total'].sum().sort_values(ascending=False).index.tolist()

# Crear el gráfico de barras con hue y especificar el orden de las provincias en el eje y
plt.figure(figsize=(10, 8))
sns.barplot(x='Total', y='Provincia', hue='Periodo', data=conexiones_totales_provincia, order=provincias_ordenadas)

# Configurar las etiquetas
plt.xlabel('Cantidad de conexiones')
plt.ylabel('Provincia')
plt.title('Cantidad Total de Conexiones por Provincia (2018 al 2022 - Q1)')
plt.legend(title='Periodo')

# Mostrar el gráfico
plt.show()
```

##### **Analisis: Cantidad de Conexiones por Provincia (2018 al 2022 - Q1)**
* Aqui podemos ver la cantidad de conexiones por provincia en el primer trimestre de los 5 ultimos años. Hice el analis asi debido a que en el 2022 solo tenemos acceso a los datos de este trimestre, de esta forma se tienen un analisis más comparable a lo largo de cada año.

* Se puede ver que ha ido incrementando en todas las provincias la cantidad de conexiones a internet por año. Siendo el crecimiento mas notable, de Buenos Aires, Capital Federal, Cordoba y Santa Fe. Esto se puede deber a que la población y el area territorial es mayor en estas regiones y no es una metrica tan real para ver el progreso en cada una de las provincias.

```python
# Reemplazo las letras "í" con tilde por la letra "i" sin tilde en la columna 'Provincia'del df conexiones_totales_provincia para poder hacer el merge.

conexiones_totales_provincia['Provincia'] = conexiones_totales_provincia['Provincia'].str.replace('é', 'e')
conexiones_totales_provincia['Provincia'] = conexiones_totales_provincia['Provincia'].str.replace('á', 'a')
conexiones_totales_provincia['Provincia'] = conexiones_totales_provincia['Provincia'].str.replace('í', 'i')
conexiones_totales_provincia['Provincia'] = conexiones_totales_provincia['Provincia'].str.replace('ó', 'o')
```

#### Grafico de barras: Crecimiento del Acceso de Conexiones a Internet en las Provincias con menos del 20% de Acceso por Población: Análisis Q1-2018 al 2022'

El interes esta en las provincias con menos del 20%, creo un nuevo dataframe que las contenga.

```python
# Crear una lista de provincias específicas
provincias_seleccionadas = ['Tucuman', 'Jujuy', 'Salta','Mendoza','Misiones','Catamarca','Santa Cruz','Santiago Del Estero','Corrientes', 'Chaco', 'San Juan', 'Formosa']

# Filtrar el DataFrame para incluir solo las provincias seleccionadas
conexiones_totales_provincia_menores = conexiones_totales_provincia[conexiones_totales_provincia['Provincia'].isin(provincias_seleccionadas)]

# Crear el gráfico de barras con hue utilizando solo las provincias seleccionadas
plt.figure(figsize=(10, 8))
sns.barplot(x='Porcentaje de Conexiones Totales', y='Provincia', hue='Periodo', data=conexiones_totales_provincia_menores, order=provincias_seleccionadas)

# Configurar las etiquetas
plt.xlabel('%')
plt.ylabel('Provincia')
plt.title('Porcentaje de la Población por Provincia con Accesos de Conexión a Internet desde 1 kbps a +30 Mbps: Análisis')
plt.legend(title='Periodo')

# Mostrar el gráfico
plt.show()
```


#### Grafico de barras: Crecimiento Porcentual del Acceso de Conexiones a Internet en las Provincias con menos del 20% de Acceso por Población: Análisis del 2018 al 2022'

```python
# Filtrar los datos para los periodos 2018-I y 2022-I
df_filtrado = conexiones_totales_provincia_menores[(conexiones_totales_provincia_menores['Periodo'] == '2018-Q1') | (conexiones_totales_provincia_menores['Periodo'] == '2022-Q1')]

# Pivoteo de la tabla
df_pivot = df_filtrado.pivot(index='Provincia', columns='Periodo', values='Porcentaje de Conexiones Totales')

# Calcular el crecimiento porcentual
df_pivot['Crecimiento'] = df_pivot['2022-Q1'] - df_pivot['2018-Q1']

# Ordenar el DataFrame por el crecimiento porcentual en orden descendente
df_pivot.sort_values('Crecimiento', ascending=False, inplace=True)

# Lista de tonos de azul
colors = ['#0A224E', '#133867', '#1C4B80', '#245F99', '#2D74B2', '#368DCB', '#41A6E4', '#5EB3E8', '#7CBFEA', '#9ADBEF', '#B8E7F3', '#D6F3F8']


# Crear el gráfico de barras horizontales con la paleta de colores personalizada
plt.figure(figsize=(10, 6))
sns.barplot(x='Crecimiento', y=df_pivot.index, data=df_pivot, orient='h', palette=colors)

# Configurar las etiquetas
plt.xlabel('Crecimiento Porcentual')
plt.ylabel('Provincia')
plt.title('Crecimiento Porcentual de Conexiones por Provincia por Población (Q1 2018 - 2022)')


# Mostrar los valores en cada barra como porcentaje
for i, value in enumerate(df_pivot['Crecimiento']):
    plt.text(value, i, f' {value*100:.2f}%', ha='left', va='center')

plt.show()
```

```python
# Obtener el número de provincias por año con velocidad mayor o igual a 30 Mbps
provincias_con_velocidad_mayor_30 = velocidad_media_provincia[velocidad_media_provincia['Mbps (Media de bajada)'] >= 30].groupby('Año')['Provincia'].nunique()
provincias_con_velocidad_mayor_30
```

```python
# Calcular el porcentaje de provincias con velocidad mayor a 25 Mbps para cada año
total_provincias = 24  # Considerando que hay 24 provincias en total
porcentaje_provincias_mayor_30 = (provincias_con_velocidad_mayor_30 / total_provincias) * 100

# Crear una nueva columna en el DataFrame con los porcentajes calculados
velocidad_media_provincia['Porcentaje_Provincias_Mayor_30'] = velocidad_media_provincia['Año'].map(porcentaje_provincias_mayor_30)

# LLeno los nulos con 0 (años en los que el porcentaje haya sido nulo)
velocidad_media_provincia.fillna(0, inplace=True)

velocidad_media_provincia
```

#### Grafico de Barras: Crecimiento Porcentual de las provincias con más de 30 Mbps de velocidad media de bajada por Año

```python
# Filtrar los datos solo para las columnas 'Year' y 'Porcentaje_Provincias_Mayor_30'
data = velocidad_media_provincia[['Año', 'Porcentaje_Provincias_Mayor_30']].copy()

# Reordenar los años de manera ascendente
data.sort_values('Año', inplace=True)

# Eliminar las filas duplicadas basadas en la columna 'Year'
data.drop_duplicates(subset='Año', inplace=True)

# Calcular el crecimiento porcentual entre años consecutivos
data['Crecimiento_Porcentual'] = data['Porcentaje_Provincias_Mayor_30'].diff()

# Eliminar la última fila que contendría un NaN en la columna de crecimiento porcentual
data.dropna(inplace=True)

# Crear una columna para el intervalo de tiempo
data['Intervalo_Tiempo'] = (data['Año']-1).astype(str) + '-' + (data['Año'] ).astype(str)

# Crear el gráfico de barras para el crecimiento porcentual por año
plt.figure(figsize=(8, 4))
bars = plt.bar(data['Intervalo_Tiempo'], data['Crecimiento_Porcentual'], color='skyblue', edgecolor='black')

# Agregar etiquetas de valor en las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}%', ha='center', va='bottom')

# Configurar los ejes y el título del gráfico
plt.xlabel('Año')
plt.ylabel('Crecimiento Porcentual')
plt.title('Crecimiento Porcentual de las provincias con más de 30 Mbps de velocidad media de bajada por Año')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Mostrar el gráfico
plt.show()
```
